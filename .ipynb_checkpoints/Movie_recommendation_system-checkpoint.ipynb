```python
# needed only name part
# have to convert to a lsit like ['Action','Adventure','Science Fiction']
```

```python
# helper func
def convert_genre(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
convert_genre('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# here data is too large
# so the object contains set of dictionaries
# we'll take first 3 dictionary --> then three names(actors)
```

```python
def convert_cast(obj):
    L = []
    cnt=0
    for i in ast.literal_eval(obj):
        if(cnt == 3):
            break
        L.append(i['name'])
        cnt+=1     
        
    return L
```

```python
movies['cast'].apply(convert_cast)
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
# will take only those sec where job = director
```

```python
def convert_crew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if(i['job'] == 'Director'):
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'].apply(convert_crew)
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
```

```python
# text vectorization
# finding similaraties between the tags
```

```python
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
print(vectors)
```

```python
vectors[0]
```

```python
# stemming --> removing similar types of words into main word
```

```python
ps = PorterStemmer()
```

```python
def stemming(text):
    L = []
    for i in text.split():
        L.append(ps.stem(i))
    #list to string as we r working with string
    string = " ".join(L)
    return string
```

```python
stemming('dancing')
```

```python
stemming('dance')
```

```python
new_df['tags'].apply(stemming)
```

```python
new_df['tags'] = new_df['tags'].apply(stemming)
```

```python
# recommendation creation
# index -> find index in similarity --> sort --> choose closest
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True , key = lambda x:x[1])[1:11]
    
    for i in movies_list:
        # print(i[0])  --> index value of touple
        print(new_df['title'][i[0]])
    
```

```python
recommend('Avatar')
```
